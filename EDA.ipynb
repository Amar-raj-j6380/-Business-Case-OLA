```python
column_aliases = {
    'Current_date': 'Current_date',
    'MMM-YY': 'Total_Months_Worked',
    'Grade':'Max_Grade',
    'Income': 'Income_Total',
    'Age': 'Age_Max',
    'Total Business Value': 'Total_Business_Value',
    'Quarterly Rating': 'Quarterly_Rating_Mean',
    'Churn_User' : 'Churn_User'
}


df_agg = df.groupby(['Driver_ID', 'Gender', 'City', 'Education_Level', 'Joining Designation'])\
            .agg({'Current_date':'max', 'MMM-YY':'count',  'Grade':'max','Income': 'sum', 'Age':'max', 'Total Business Value': 'sum', 'Quarterly Rating': 'mean',
                  'Churn_User':'max'})\
            .rename(columns=column_aliases).reset_index(inplace=True)
```

```python
# Example DataFrame
data = {'Driver': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
        'Quarter': [1, 2, 3, 4, 1, 2, 3, 4],
        'Rating': [80, 85, 90, 85, 70, 75, 80, 85]}
df1 = pd.DataFrame(data)

# Sort the DataFrame by Driver and Quarter
df1 = df1.sort_values(by=['Driver', 'Quarter'])

# Calculate the difference in rating between consecutive quarters
df1['Rating_Difference'] = df1.groupby('Driver')['Rating'].diff()

# Create a column to indicate if rating has increased (1) or not (0)
df1['Increased'] = (df1['Rating_Difference'] > 0).astype(int)

print(df1)
```
